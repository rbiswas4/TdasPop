```python
from tdaspop import SersicSamples
import tdaspop
print(tdaspop.__version__)
```

```python
import numpy as np
```

```python
ss = SersicSamples(np.random.RandomState(1))
```

```python
num_samples = 1000000
```

```python
halflightRadius = np.ones(num_samples)
```

```python
s1 = ss.sampleRadius(halflightRadius, numSamples=len(halflightRadius), sersicIndex=1)
s4 = ss.sampleRadius(halflightRadius, numSamples=len(halflightRadius), sersicIndex=4)
```

```python
import seaborn as sns
sns.set_style('whitegrid')
sns.set_context('talk')
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
_ = ax.hist(s1, bins=np.arange(0, 100, 0.1), histtype='step', normed=1, label='s1')
_ = ax.hist(s4, bins=np.arange(0, 100, 0.1), histtype='step', normed=1, label='s4')
_ = ax.legend(loc='best')
```

## Let us look at it more closely

```python
fig, ax = plt.subplots()
_ = ax.hist(s1, bins=np.arange(0, 100, 0.1), histtype='step', density=1, label='s1')
_ = ax.hist(s4, bins=np.arange(0, 100, 0.1), histtype='step', density=1, label='s4')
_ = ax.legend(loc='best')
ax.set_xlim(0, 10)
```

## Angles

```python
angles = ss.sampleAngles(np.ones(num_samples), np.ones(num_samples), num_samples)
```

```python
fig, ax = plt.subplots()
_ = ax.hist(angles, bins=np.arange(0, 400, 5), histtype='step', density=1, label='s1')
```

### Angles in different ellipticities

The angles are with respect to the major axis (which is 0., assuming a is made the larger value)

```python
a = 1.2
b = 0.8
angles = ss.sampleAngles(np.ones(num_samples)*a, np.ones(num_samples)*b, num_samples)
```

```python
fig, ax = plt.subplots()
_ = ax.hist(angles, bins=np.arange(0, 400, 5), histtype='step', density=1, label='s1')
```
